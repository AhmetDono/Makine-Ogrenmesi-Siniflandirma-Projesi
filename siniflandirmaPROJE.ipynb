```python
y = df["Type"].values
x = df.drop("Type",axis=1).values
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler=MinMaxScaler()
```

```python
scaler.fit(x_train)
```

```python
x_train=scaler.transform(x_train)
```

```python
x_test=scaler.transform(x_test)
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation,Dropout
from tensorflow.keras.callbacks import EarlyStopping
```

```python
x_train.shape
```

```python
model=Sequential()
#ufak kural kaç kolon varsa ilk layera o kadar nöron koyulur
#aralarada 1 ve 30 arasında nöron konur
model.add(Dense(30,activation="relu"))
model.add(Dense(15,activation="relu"))
model.add(Dense(15,activation="relu"))
#final output
model.add(Dense(1,activation="sigmoid"))
#birleştirici loss türleri
model.compile(optimizer='adam',loss="binary_crossentropy")
```

```python
                                        #doğrulama datası
model.fit(x=x_train,y=y_train,epochs=700,validation_data=(x_test,y_test),verbose=1)
```

```python
modelKaybi=pd.DataFrame(model.history.history)
```

```python
modelKaybi.plot()
```

```python
model=Sequential()
#ufak kural kaç kolon varsa ilk layera o kadar nöron koyulur
#aralarada 1 ve 30 arasında nöron konur
model.add(Dense(30,activation="relu"))
model.add(Dense(15,activation="relu"))
model.add(Dense(15,activation="relu"))
#final output
model.add(Dense(1,activation="sigmoid"))
#birleştirici loss türleri
model.compile(optimizer='adam',loss="binary_crossentropy")
```

```python
earlyStopping=EarlyStopping(monitor="val_loss",mode="min",verbose=1,patience=25)
```

```python
model.fit(x=x_train,y=y_train,epochs=700,validation_data=(x_test,y_test),verbose=1,callbacks=[earlyStopping])
```

```python
modelKaybi=pd.DataFrame(model.history.history)
```

```python
modelKaybi.plot()
```

```python
model=Sequential()
#ufak kural kaç kolon varsa ilk layera o kadar nöron koyulur
#aralarada 1 ve 30 arasında nöron konur
model.add(Dense(30,activation="relu"))
model.add(Dropout(0.5))
model.add(Dense(15,activation="relu"))
model.add(Dropout(0.5))
model.add(Dense(15,activation="relu"))
model.add(Dropout(0.5))
#final output
model.add(Dense(1,activation="sigmoid"))
#birleştirici loss türleri
model.compile(optimizer='adam',loss="binary_crossentropy")
```

```python
model.fit(x=x_train,y=y_train,epochs=700,validation_data=(x_test,y_test),verbose=1,callbacks=[earlyStopping])
```

```python
modelKaybi=pd.DataFrame(model.history.history)
```

```python
modelKaybi.plot()
```

```python
tahminlerimiz=model.predict_classes(x_test)
```

```python
tahminlerimiz
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
print(classification_report(y_test,tahminlerimiz))
```

```python
print(confusion_matrix(y_test,tahminlerimiz))
#5 değer yanlış bulmuş demek
```
